## CALCUL D´UN MOYENNE                             

```python
marta= [10, 15, 11, 18, 7, 12, 13, 16, 18]
tiago= [8, 11, 12, 14, 10, 9, 15, 12, 11,13,18]
```

<br>*Complétez la fonction `moyenne(liste)` qui permet de calculer la moyenne d´un élève arrondie au dixième.*

```python
# code
def moyenne(liste):
    # on crée une variable pour calculer la somme des notes
    somme=0
    # on parcourt les indices de la liste pouir calculer la somme des notes
    for i in range(len(liste)):
        somme+=liste[i]
    # on retourne la moyenne arrondie au dixième
    return round(somme/len(liste),1)
```

<br>Affichez les moyennes de Marta et Tiago:

```python
# Marta
moyenne(marta)
```

```python
# Tiago
moyenne(tiago)
```

<br>*Tiago a obtenu un **14** et Marta un **16***. Comment ajouter ces deux nouvelles notes aux listes?

```python
# code
marta.append(16)
tiago.append(14)
```

<br>*Le professeur décide de cocher la case "enlever la plus mauvaise note". Complétez la fonction `indexMini` puis affichez la moyenne corrigée de Marta et Tiago.*

Consultez le fonctionnement de l´algorithme sur mon site: <br>
http://www.fredpeuriere.com/NSI/algos-1/algorithmes-illustres/min.html

```python
# la fonction:
def indexMini(liste):  
    """détermination de l´indice 
    de la plus mauvaise note"""
    # on commence par initialiser l´indice du minimium au premier élément  
    min=0
    # dans la boucle, on détermine l´indice du nouveau minimum
    # si on trouve un élément plus petit
    for i in range(len(liste)):
        if liste[i]<liste[min]:
            min=i
    return min
```

```python
# Indice de la plus mauvaise note de MARTA
indexMini(marta)
```

```python
# Indice de la plus mauvaise note de Tiago
indexMini(tiago)
```

```python
# Moyenne corrigée de MARTA

# On utilise la méthode pop(i) qui supprime la valeur à l´indice i
# MARTA:
marta.pop(indexMini(marta))
# On affiche la nouvelle liste de notes:
print(marta)
# On affiche la nouvelle moyenne:
print(moyenne (marta))
```

```python
# Moyenne corrigée de TIAGO

tiago.pop(indexMini(tiago))
# On affiche la nouvelle liste de notes:
print(tiago)
# On affiche la nouvelle moyenne:
print(moyenne (tiago))
```

<br>*Un carré est magique lorsque la somme de ces lignes, colonnes et diagonales sont égales*<br>

```python
carre3=[
    [2,7,6],
    [9,5,1],
    [4,3,8],
]

carre4=[
    [4,5,11,14],
    [15,10,8,1],
    [6,3,13,12],
    [9,16,2,7]
]
```


> Quelle est la valeur de `len(carre4)` ? <br>
> Quelle est la valeur de `carre3[0][2]`? <br>
> Quelle instruction permet de récupérer la valeur 3 de `carre4` ?
<img src="http://www.fredpeuriere.com/NSI/5/carre.png" width='300'>

```python
# tests
print(len(carre4))
print(carre3[0][2])
# affichage du 3 de carre4 (3e ligne, 2e colonne):
print(carre4[2][1])
```

```python
def sommeLigne(carre,n):
    """carre est un tableau carré de nombres
    n est un nombre entier"""
    somme=0
    for i in carre[n]:
        somme+=i
    return somme
```

<br>Que vaut `sommeLigne(carre4, 2)` ? À quoi sert cette fonction ?

```python
# On calcule ici la somme des nombres sur la troisième ligne (indice 2) de carre4
# La fonction calcule la somme des valeurs d´un carré sur une ligne déterminée
sommeLigne(carre4, 2)
```

```python
# test avec le carre3, première ligne
sommeLigne(carre3, 0)
```

<br>Écrire le code d’une fonction `sommeDesLignes` qui prend un carré en paramètre et qui vérifie que les sommes des nombres de chaque ligne sont égales.

```python
# réponse

def sommeDesLignes(carre):
    # on calcule la somme des nombres en première ligne
    somme = sommeLigne(carre, 0)
    # on parcourt les autres lignes
    for i in range(1, len(carre)):
        # Si la somme calculée est différente de la première
        # On retourne False
        if somme != sommeLigne(carre,i):
            return False
    return True
```

```python
# test avec le carre3
sommeDesLignes(carre3)
```

```python
# test avec le carre4
sommeDesLignes(carre4)
```

<br>Proposer le code d’une fonction `sommeColonne` qui prend un carré en paramètre, ainsi que le numéro d’une colonne, et qui renvoie la somme des nombres de cette colonne.<br>
<img src="http://www.fredpeuriere.com/NSI/5/carre.png" width='300'>

```python
# réponse
def sommeColonne(carre,n):
    somme= 0
    for i in range(len(carre)):
        somme+=carre[i][n]
    return somme
```

```python
# test avec le carre4, deuxième colonne
sommeColonne(carre4,1)
```

<br>Pour info, la somme des diagonales:

```python
def sommeDiagonales (carre):
    somme1,somme2 = 0,0
    for i in range(len(carre)):
        somme1+=carre[i][i]
        somme2+=carre[i][len(carre)-1-i]
    return somme1,somme2
```


## DICTIONNAIRES: AU ZOO
Au jardim zoologico de Lisbonne, il y a 5 éléphants d’Asie, 17 écureuils d’Asie, 2 pandas d’Asie… On représente cet inventaire à l’aide d’un dictionnaire, de la façon suivante :

```python
zoo_Lisboa={ 'éléphant': ('Asie', 5),  'écureuil': ('Asie', 17), 
            'panda': ('Asie', 2), 'hippopotame': ('Afrique', 7),  
            'girafe': ('Afrique', 4)}
```

On représente de la même façon le zoo de Porto: 

```python
zoo_Porto = { 'ours': ('Europe', 4),  'tigre': ('Asie', 7), 
             'girafe': ('Afrique', 11), 'hippopotame': ('Afrique', 3)}  
```

On souhaite créer la fonction `plus_grand_nombre` qui prend un zoo en paramètre et qui renvoie le nom de l’animal le plus représenté dans ce zoo. Par exemple :
> plus_grand_nombre(zoo_Porto) --> 'girafe'<br>
> plus_grand_nombre(zoo_Lisboa) --> 'écureuil'


a. Quel type de boucle peut-on envisager pour le code de cette fonction ? <br><br>
<span>☐ `for i in dico.keys()`       ☐ `for i in dico.values()`    <br>   ☐ `for (i,j) in dico.items()`       ☐ `Aucune boucle.`</span>


```python
# On itère sur les items
# i représente les animaux --> les clés
# j représente le tuple (continent,nombre) ---> les valeurs
# Exemple à Lisbone:
for i,j in zoo_Lisboa.items():
    print(i,j)
```

  b. Écrire et tester le code de cette fonction. 

```python
# fonction: on procède comme la recherche du minimum
# on cherche ici le maximum:
def plus_grand_nombre(dico):
    # on initialise le nom et le nombre d´animaux
    animal=''
    nombre=0
    for i,j in dico.items():
        # On accède au nombre par j[1]
        if j[1]>nombre:
            # mise à jour du nombre d´animal:
            nombre=j[1]
            # On accède au nom de l´animal par i:
            animal=i
    return animal
```

```python
# test de la fonction
print(plus_grand_nombre(zoo_Porto))
print(plus_grand_nombre(zoo_Lisboa))
```

a. Quel type de boucle peut-on envisager pour le code de cette fonction ? <br><br>
<span>☐ `for i in dico.keys()`       ☐ `for i in dico.values()`    <br>   ☐ `for (i,j) in dico.items()`       ☐ `Aucune boucle.`</span>


```python
# On itère sur les valeurs qui sont des tuples
# contenant le continent et le nombre d´animaux.
# Exemple à Porto:
for i in zoo_Porto.values():
    print(i)
```

  b. Écrire et tester le code de cette fonction. 

```python
# fonction
def nombre_total(dico,continent):
    # on initialise le nombre d´animaux
    nombre=0
    for i in dico.values():
        # On accède au continent par i[0]
        if i[0]==continent:
            # On accède au nombre par i[1]
            nombre+=i[1]
    return nombre
```
